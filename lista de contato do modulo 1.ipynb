```python
def adicionar_contato(contatos, nome, telefone, email, favorito=False):
    # Adiciona um novo contato ao dicionário 'contatos'.
    # O contato é armazenado com 'nome' como chave e um dicionário com telefone, email e status de favorito como valor.
    contatos[nome] = {"telefone": telefone, "email": email, "favorito": favorito}



def listar_contatos(contatos):
    # Itera sobre todos os contatos no dicionário 'contatos' e imprime suas informações.
    for nome, info in contatos.items():
        print(f"Nome: {nome}, Telefone: {info['telefone']}, Email: {info['email']}, Favorito: {'Sim' if info['favorito'] else 'Não'}")
```

```python
def menu():
    # Cria um dicionário vazio 'contatos' para armazenar os dados dos contatos.
    contatos = {}

    # Loop infinito para manter o programa rodando até que o usuário decida sair.
    while True:
        # Exibe as opções do menu.
        print("\nAgenda de Contatos")
        print("1. Adicionar contato")
        print("2. Listar contatos")
        print("3. Sair")

        # Solicita ao usuário para escolher uma opção.
        escolha = input("Escolha uma opção: ")

        # Avalia a escolha do usuário e chama a função apropriada.
        if escolha == "1":
            # Solicita informações do contato ao usuário.
            nome = input("Nome: ")
            telefone = input("Telefone: ")
            email = input("Email: ")
            # Adiciona o contato ao dicionário 'contatos'.
            adicionar_contato(contatos, nome, telefone, email)
        elif escolha == "2":
            # Lista todos os contatos.
            listar_contatos(contatos)
        elif escolha == "3":
            # Sai do programa.
            break
        else:
            # Informa ao usuário que a opção escolhida é inválida.
            print("Opção inválida. Tente novamente.")

# A linha abaixo está comentada para evitar a execução automática no ambiente atual.
```

```python
def editar_contato(contatos, nome):
    # Verifica se o contato existe no dicionário.
    if nome in contatos:
        print(f"Editando contato: {nome}")
        # Solicita novos valores para cada atributo do contato.
        novo_telefone = input("Novo Telefone (deixe em branco para manter o atual): ")
        novo_email = input("Novo Email (deixe em branco para manter o atual): ")
        novo_favorito = input("Marcar como favorito? (s/n): ")

        # Atualiza apenas os campos que o usuário deseja alterar.
        if novo_telefone:
            contatos[nome]['telefone'] = novo_telefone
        if novo_email:
            contatos[nome]['email'] = novo_email
        if novo_favorito.lower() in ['s', 'sim']:
            contatos[nome]['favorito'] = True
        elif novo_favorito.lower() in ['n', 'não', 'nao']:
            contatos[nome]['favorito'] = False
    else:
        print("Contato não encontrado.")

def menu():
    contatos = {}

    while True:
        print("\nAgenda de Contatos")
        print("1. Adicionar contato")
        print("2. Listar contatos")
        print("3. Editar contato")
        print("4. Sair")
        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            nome = input("Nome: ")
            telefone = input("Telefone: ")
            email = input("Email: ")
            adicionar_contato(contatos, nome, telefone, email)
        elif escolha == "2":
            listar_contatos(contatos)
        elif escolha == "3":
            nome_para_editar = input("Digite o nome do contato a ser editado: ")
            editar_contato(contatos, nome_para_editar)
        elif escolha == "4":
            break
        else:
            print("Opção inválida. Tente novamente.")

# menu()
```

```python
def alterar_favorito(contatos, nome):
    # Verifica se o contato existe no dicionário.
    if nome in contatos:
        # Alterna o status de favorito do contato.
        contatos[nome]['favorito'] = not contatos[nome]['favorito']
        status_favorito = "Favorito" if contatos[nome]['favorito'] else "Não favorito"
        print(f"Contato '{nome}' agora está marcado como: {status_favorito}")
    else:
        print("Contato não encontrado.")

def menu():
    contatos = {}

    while True:
        print("\nAgenda de Contatos")
        print("1. Adicionar contato")
        print("2. Listar contatos")
        print("3. Editar contato")
        print("4. Marcar/Desmarcar como favorito")
        print("5. Sair")
        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            nome = input("Nome: ")
            telefone = input("Telefone: ")
            email = input("Email: ")
            adicionar_contato(contatos, nome, telefone, email)
        elif escolha == "2":
            listar_contatos(contatos)
        elif escolha == "3":
            nome_para_editar = input("Digite o nome do contato a ser editado: ")
            editar_contato(contatos, nome_para_editar)
        elif escolha == "4":
            nome_para_alterar_favorito = input("Digite o nome do contato para alterar o favorito: ")
            alterar_favorito(contatos, nome_para_alterar_favorito)
        elif escolha == "5":
            break
        else:
            print("Opção inválida. Tente novamente.")

#
```

```python
def listar_favoritos(contatos):
    # Verifica se existem contatos favoritos.
    tem_favoritos = any(info['favorito'] for info in contatos.values())
    if not tem_favoritos:
        print("Não há contatos favoritos.")
        return

    # Itera sobre todos os contatos e imprime apenas os marcados como favoritos.
    for nome, info in contatos.items():
        if info['favorito']:
            print(f"Nome: {nome}, Telefone: {info['telefone']}, Email: {info['email']}")

def menu():
    contatos = {}

    while True:
        print("\nAgenda de Contatos")
        print("1. Adicionar contato")
        print("2. Listar contatos")
        print("3. Editar contato")
        print("4. Marcar/Desmarcar como favorito")
        print("5. Listar favoritos")
        print("6. Sair")
        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            nome = input("Nome: ")
            telefone = input("Telefone: ")
            email = input("Email: ")
            adicionar_contato(contatos, nome, telefone, email)
        elif escolha == "2":
            listar_contatos(contatos)
        elif escolha == "3":
            nome_para_editar = input("Digite o nome do contato a ser editado: ")
            editar_contato(contatos, nome_para_editar)
        elif escolha == "4":
            nome_para_alterar_favorito = input("Digite o nome do contato para alterar o favorito: ")
            alterar_favorito(contatos, nome_para_alterar_favorito)
        elif escolha == "5":
            listar_favoritos(contatos)
        elif escolha == "6":
            break
        else:
            print("Opção inválida. Tente novamente.")


```

```python
def apagar_contato(contatos, nome):
    # Verifica se o contato existe no dicionário.
    if nome in contatos:
        # Remove o contato do dicionário.
        del contatos[nome]
        print(f"Contato '{nome}' apagado com sucesso.")
    else:
        print("Contato não encontrado.")

def menu():
    contatos = {}

    while True:
        print("\nAgenda de Contatos")
        print("1. Adicionar contato")
        print("2. Listar contatos")
        print("3. Editar contato")
        print("4. Marcar/Desmarcar como favorito")
        print("5. Listar favoritos")
        print("6. Apagar contato")
        print("7. Sair")
        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            nome = input("Nome: ")
            telefone = input("Telefone: ")
            email = input("Email: ")
            adicionar_contato(contatos, nome, telefone, email)
        elif escolha == "2":
            listar_contatos(contatos)
        elif escolha == "3":
            nome_para_editar = input("Digite o nome do contato a ser editado: ")
            editar_contato(contatos, nome_para_editar)
        elif escolha == "4":
            nome_para_alterar_favorito = input("Digite o nome do contato para alterar o favorito: ")
            alterar_favorito(contatos, nome_para_alterar_favorito)
        elif escolha == "5":
            listar_favoritos(contatos)
        elif escolha == "6":
            nome_para_apagar = input("Digite o nome do contato a ser apagado: ")
            apagar_contato(contatos, nome_para_apagar)
        elif escolha == "7":
            break
        else:
            print("Opção inválida. Tente novamente.")
menu()
```
